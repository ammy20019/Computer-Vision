```python
#animation
#translation of images
image1=cv2.imread("C:/Users/Admin/Pictures/cartoon.jpg")
row,col,ch=image1.shape
x=0
y=0
while(True):
    T=np.float64([[1,0,x],[0,1,y]])
    image2=cv2.warpAffine(image1,T,(col,row))
    cv2.imshow("frame1",image1)
    cv2.imshow("frame2",image2)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
    x=x+.1
cv2.destroyAllWindows()
```

```python
#rotation of image
image1=cv2.imread("C:/Users/Admin/Pictures/cartoon.jpg")
row,col,ch=image1.shape
angle=0

while(True):
    if angle==360:
        angle=0
    
    R=cv2.getRotationMatrix2D((row/2,col/2),angle,0.6)

    image2=cv2.warpAffine(image1,R,(col,row))

    cv2.imshow("frame1",image1)
    cv2.imshow("frame2",image2)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
    angle=angle+1
cv2.destroyAllWindows()
```

### blurring

```python
import cv2
import numpy as np

image1=cv2.imread("C:/Users/Admin/Pictures/cartoon.jpg")

kernel_5x5=(np.ones((5,5),np.uint8))/25
kernel_9x9=(np.ones((9,9),np.uint8))/81

image2=cv2.filter2D(image1, -1,kernel_5x5)
image3=cv2.filter2D(image1, -1,kernel_9x9)

image4=cv2.blur(image1,(5,5))
image5=cv2.GaussianBlur(image1,(5,5),0)
image6=cv2.medianBlur(image1,5)
image7=cv2.bilateralFilter(image1,9,75,75)

cv2.imshow("original",image1)
cv2.imshow("filter2d",image2)
cv2.imshow("filter2d2",image3)
cv2.imshow("Blur",image1)
cv2.imshow("Gaussian",image2)
cv2.imshow("median",image3)


cv2.waitKey()
cv2.destroyAllWindows()
```
